```python
character_remove = [',','₹','%']
columns = ['discounted_price','actual_price','discount_percentage','rating_count']
```

```python
for ch in character_remove:
    for cl in columns:
        data[cl] = data[cl].str.replace(ch,'')
```

```python
for cl in columns:
    data[cl] = data[cl].astype('float64')
```

```python
data.info()
```

```python
data[data['rating'].str.contains('|', regex=False)]
```

```python
data['rating'] = data['rating'].str.replace('|','4').astype('float64')
```

```python
data['category'].value_counts()
```

```python
data['category'] = data.category.apply(lambda x:x.split('|')[0].replace('&',' & '))
```

```python
data['category'].value_counts()
```

```python
sns.countplot(data = data, y = 'category')
```

```python
plt.figure(figsize=(8,6))
plt.title('Products Demand in Price Range')
sns.stripplot(data = data, x = 'actual_price', y = 'category', hue='category')
```

```python
plt.figure(figsize=(10,8))
plt.title('Ratings of Products In Categories')
sns.stripplot(data = data, x = 'rating', y = 'category', hue= 'category')
```

```python
discount_per = data.groupby('category')['discount_percentage'].apply(lambda x:x.mode()).sort_values(ascending = False).reset_index()
discount_per
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
plt.title('Top Discount Percentage Categories')
sns.barplot(ax=ax,data = discount_per, y =discount_per.category,  x = discount_per.discount_percentage,palette='Blues')
ax.bar_label(ax.containers[0],fmt='%.0f%%')
```

```python
data['sales'] = data.discounted_price*data.rating_count
data['sales']
```

```python
total_sales = data.groupby('category')['sales'].sum().sort_values(ascending = False)
```

```python
total_sales
```

```python
import math
```

```python
import math

millnames = ['',' Thousand',' Million',' Billion',' Trillion']

def millify(n):
    n = float(n)
    millidx = max(0,min(len(millnames)-1,
                        int(math.floor(0 if n == 0 else math.log10(abs(n))/3))))

    return '{:.0f}{}'.format(n / 10**(3 * millidx), millnames[millidx])
```

```python
values = []
for x in total_sales.values:
    values.append(millify(x))
    print(millify(x))
```

```python
fig,ax = plt.subplots(figsize=(8,6))
plt.title('Total Sales of Categories')
sns.barplot(data = data, y = total_sales.index, x = total_sales.values)
ax.set_ylabel('Categories')
ax.bar_label(ax.containers[0], labels=values)
plt.show()
```
